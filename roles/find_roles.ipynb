```python
import csv
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("company_roles.csv")
df
```

```python
temp_roles = pd.read_csv("roles.csv")
temp_roles
```

```python
def clean_up(data):
    res = data.replace('"', '')
    res = res.replace('\'', '')
    res = res.replace(',', '')
    res = res.replace('  ', '')
    res = res.upper()
    return res
```

```python
roles1 = df['ROLE_NAME'].astype(str).apply(clean_up).unique()
roles2 = temp_roles['NAME'].astype(str).unique()
all_roles = np.concatenate((roles1, roles2), axis=None)
all_roles = np.unique(all_roles)
all_role_df = pd.DataFrame(all_roles, columns = ['NAME'])
all_role_df = all_role_df.reset_index(drop=True)
all_role_df.index.name = 'ROLE_ID'
all_role_df.to_csv("roles_final.csv")
```
