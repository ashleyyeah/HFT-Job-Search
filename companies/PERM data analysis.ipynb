```python
def normalize(data):
    res = data.replace(',', '')
    res = res.replace('.', '')
    res = res.upper()
    return res
```

```python
def state_to_code(state_name):
    state_name = ''.join([i for i in state_name if not i.isdigit()])
    state = us.states.lookup(state_name)
    if state == None:
        return state_name
    return state.abbr
```
